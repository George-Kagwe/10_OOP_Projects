```r
#regression/aka ordinary least squares
#perform a simple linear regression with R
#plot time series data
#fit a linear model to a set of ordered pairs
#use of Mauna Loa co2 dataset

# plot(co2,main='Atmospheric CO2 Concetration')

#the idea behind linear regression is that you have a response variable in this case co2 concetration
#of the ith observation which may be denoted by the random variable Yi
#this response depends upon the explanatory varaible xi in a linear way with some noise added as
# Yi= linear model plus noise =(B0 +B1xi)+Ei
#the error term is used to
#measure error
#lack of knowledge of other important influences

#-------------------------------------
#if you want to make inferences, we have vanilla/often reosonable assumptions
# -- The errors are normally distributed and on average zero
# -- the errors all have the same variance ( they are homoscedastic) 
# -- the errors are unrelated to each other they are independed across variations

# Develop your linear model:
co2.linear.model=lm(co2 ~ time(co2))

plot(co2, main = 'Atmospheric Co2 Concentration with Fitted line')

abline(co2.linear.model)
```

```r
#part II
#assesing the residuals 
#assumption is that they are normally disributed
co2.lm=lm(co2~time(co2))
co2.residuals=resid(co2.lm)
hist(co2.residuals,main='Histogram of Residuals')
#histogram will not help very much in assessing normality with a small data set
```

```r
#so its best to use a quartile quartile (qqnorm)
qqnorm(co2.residuals)
qqline(co2.residuals)
```

```r
# Do a time plot of the residuals
plot(co2.residuals~time(co2))
```

```r
#Zoom in the timeplot of the residuals
plot(co2.residuals ~ time(co2),xlim=c(1960,1963),main="Zoomed Residuals")
```
